Оценим доли выбросов ряда количественных признаков в датасетах:

```python
def outliers(dataset, column):
    q1 = dataset[column].quantile(0.25) # первый квартиль
    q3 = dataset[column].quantile(0.75)  # третий квартиль
    iqr = q3 - q1  # межквартильный размах
    quant1 = q1 - 1.5*iqr  # нижняя граница выбросов
    quant3 = q3 + 1.5*iqr  # верхняя граница выбросов
    total = dataset.shape[0]  # общее количество строк в датафрейме
    emission = dataset[(dataset[column] < quant1) | \
     (dataset[column] > quant3)].shape[0]
    # количество выбросов
    emission_quant = (emission / total) * 100  # доля выбросов в столбце
    return np.round((emission_quant), decimals=2)
    # возвращаем долю выбросов в столбце
```

```python
dataset = purchases
column_list = ['quantity', 'price']
for value in column_list:
    print(f'Доля выбросов в {value}, в процентах: {outliers(dataset, value)}')
```

Доли выбросов незначительные. Отобразим данные на гистограммах и диаграммах:

```python
# Гистограмма распределения количественных признаков
def histogram(dataset, column):
    dataset[column].plot(kind='hist', bins=20, grid=True, \
                figsize=(5, 5), title=(f'Распределение признака "{column}"'))
    plt.xlabel(column)
    plt.ylabel('Количество')
    plt.show()
    return None

# Диаграмма размаха
def whiskers(dataset, column):
    dataset.boxplot(column, figsize=(5, 5))
    plt.title(f'Диаграмма размаха признака "{column}"')
    plt.ylabel('Количество')
    plt.show()
    print(dataset[column].describe().round(2))
    return None

# Диаграмма распределения категориальных признаков
def categorial(dataset, column):
    dataset[column].value_counts().plot(
    kind = 'pie',
    ylabel = '',
    autopct = '%1.1f%%',
    figsize = (5,5),
    title = f'Процентное распределение признака "{column}"')
    plt.show()
    return None
```

## Создание новых признаков и таблиц

### Признак `revenue`, таблица `revenue_agg` датасета `purchases`

### Признак `main_category`, таблица `main_cat_purchases` датасета `purchases`

В описании к данным есть следующая информация:

В таблице `purchases` есть списки идентификаторов, к каким категориям относится товар. Часто
это вложенные категории (например автотовары-аксессуары-освежители), но также может
включать в начале списка маркер распродажи или маркер женщинам/мужчинам.
Нумерация категорий сквозная для всех уровней, то есть 44 на второй позиции списка
или на третьей – это одна и та же категория. Иногда дерево категорий обновляется, поэтому
могут меняться вложенности, например ['4', '28', '44', '1594'] или ['4', '44', '1594']. Как
обработать такие случаи – можете предлагать свои варианты решения.

Для упрощения задачи оставим только главную (первую) категорию.

```python
# Функция для извлечения первого элемента из строки списка
def extract_first_element(string):
    try:
        # Преобразуем строку в список
        parsed_list = ast.literal_eval(string)
        # Проверяем, что это действительно список и берем первый элемент
        if isinstance(parsed_list, list) and parsed_list:
            return parsed_list[0]
    except (ValueError, SyntaxError):
        pass
    return None
```

```python
purchases['main_category'] = purchases['category_ids'] \
                                          .apply(extract_first_element)
```

```python
purchases.head()
```

Теперь можно создать новый признак `main_cat_purchases` (количество покупок в главной категории)

## Обучение моделей

```python
# Разделим данные на обучающую и тестовую выборки
features_train, features_test, target_train, target_test = \
                                  train_test_split(df.drop(['target'], axis=1),
                                  df['target'],
                                  test_size = TEST_SIZE,
                                  random_state = RANDOM_STATE)

features_train.shape, target_train.shape, features_test.shape, target_test.shape
```

```python
# Создадим переменные, содержащие числовые и категориальные признаки
num_features = features_train.select_dtypes(include='number').columns
cat_features = features_train.select_dtypes(exclude='number').columns
```

```python
# pipeline для OHE
ohe_pipe = Pipeline([
    ('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, \
                              strategy='most_frequent',fill_value='unknown')),
    ('ohe', OneHotEncoder(drop = 'first', categories = 'auto',
                          handle_unknown = 'ignore'))
    ])
```

```python
# pipeline для num
num_pipe = Pipeline([('imputer', SimpleImputer(strategy="constant", \
                                               fill_value=0)),
                     ('scaler', StandardScaler())
                    ])
```

```python
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, cat_features),
     ('num', num_pipe, num_features)
    ],
    remainder='passthrough'
)
```

```python
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('model', LogisticRegression(class_weight='balanced', \
                                 random_state=RANDOM_STATE))
])
```

```python
pipe_final.fit(features_train, target_train)
```

```python
# Функция для расчета метрики roc-auc
def score(model, df_features_train, df_target_train, df_features_test, \
          df_target_test, prob=True):
    # ROC-AUC на кросс-валидации
    roc_auc_train = cross_val_score(model, df_features_train, df_target_train, \
                                    scoring='roc_auc', cv=5).mean()

    # предсказанная вероятность
    target_pred = model.predict(df_features_test)

    if prob:
        probs = model.predict_proba(df_features_test)[:, 1]

    # ROC-AUC на тесте
    roc_auc_test = roc_auc_score(df_target_test, probs)

    return pd.DataFrame([roc_auc_train, roc_auc_test],
                 index=['roc_auc_train', 'roc_auc_test'],
                 columns=['score']).round(2)
```

### Модель LogisticRegression

```python
model_lr = score(pipe_final, features_train, target_train, features_test, \
                                                              target_test)
```
